```python
import requests
import json

def get_linkedin_organization_vanity_name(access_token, organization_id):
    """Retrieves the organization's vanity name using its ID."""
    url = f"https://api.linkedin.com/v2/organizations/{organization_id}"
    headers = {
        'Authorization': f'Bearer {access_token}'
    }
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an exception for bad status codes
        data = response.json()
        if 'vanityName' in data:
            return data['vanityName']
        else:
            print(f"Vanity name not found for organization ID: {organization_id}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None
    except json.JSONDecodeError:
        print("Error decoding json response")
        return None

def find_organization_id_by_name(access_token, organization_name):
    """Finds an organization's ID by its official name (not vanity name)."""
    url = f"https://api.linkedin.com/v2/organizations?q=search&keywords={organization_name}"
    headers = {'Authorization': f'Bearer {access_token}'}

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        data = response.json()

        if 'elements' in data and data['elements']:
            return data['elements'][0]['id'] #returns the first id found.
        else:
            print(f"Organization '{organization_name}' not found.")
            return None
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None
    except json.JSONDecodeError:
        print("Error decoding json response")
        return None

#Example Usage.
access_token = "AQVkD94xfeCBnY_ALiYfGDK4yzCe7q9M4i_CHKcCfJflZkbZQGixVjjWO_9NHZdjr3aeemsY7TZ7WNkX78DAdORmtpTMR_f4yOARmTQ6j8vlC2Gj4F8fRFrteQS0JxBc8NOm3BcBG5HooLV0RX3loeWOFAHdzDcOyJpvKktQpD04gPqveuoNksA9cbhunRcH-En6IGBcgTmcNZRDtO1JL-U_1TkJXmZcS7WMO0WWZTe_eLkWIGIiWYL2oNpJE75P5Vo1sunm6XbWWyrCS0DlYOO5RPOsqmRzyS6WdW6okYyc63CkDBADRpPTT3YEOJuZNBICzyvXSw_yVM8_kv-_ljaPX0LuSw"
#organization_name = "LinkedIn" #replace with the name of the organization.
organization_id = "106607818"
#organization_id = find_organization_id_by_name(access_token, organization_name)

if organization_id:
    vanity_name = get_linkedin_organization_vanity_name(access_token, organization_id)
    if vanity_name:
        print(f"The vanity name for organization is: {vanity_name}")
    else:
        print("Could not get vanity name.")
else:
    print("Could not get organization id.")
```
